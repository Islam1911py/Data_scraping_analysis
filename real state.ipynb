```python
sokhna=df.loc[df['Location'].str.contains('Ain Sukhna', case=False, na=False)]
red_sea=df.loc[df['Location'].str.contains('Red Sea', case=False, na=False)]
alexandria=df.loc[df['Location'].str.contains('alexandria', case=False, na=False)]
cairo=df.loc[df['Location'].str.contains('cairo', case=False, na=False)]
ras_elhekma=df.loc[df['Location'].str.contains('Ras Al Hekma', case=False, na=False)]
Giza=df.loc[df['Location'].str.contains('Giza', case=False, na=False)]
north_coast=df.loc[df['Location'].str.contains('North Coast', case=False, na=False) & ~df['Location'].str.contains('Ras Al Hekma', case=False, na=False)]
```

```python
import matplotlib.pyplot as plt

# احتساب عدد العقارات Off-Plan و Ready
off_plan_count = sokhna[sokhna['Completion Status'] == 'Off-Plan'].shape[0]
ready_count = sokhna[sokhna['Completion Status'] == 'Ready'].shape[0]

labels = ['Off-Plan', 'Ready']
sizes = [off_plan_count, ready_count]
colors = ['#ff9999','#66b3ff']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Off-Plan & Ready sokhna ')
plt.axis('equal')  # تأكيد على أن الدائرة دائرية

# عرض الرسم البياني
plt.show()
```

```python
# Calculate average price and count for each location
chalet_sokhna_avg_price = chalet_sokhna['Price'].astype(float).mean()
chalet_sokhna_count = chalet_sokhna.shape[0]

ras_elhekma_chalet_avg_price = ras_elhekma_chalet['Price'].astype(float).mean()
ras_elhekma_chalet_count = ras_elhekma_chalet.shape[0]

North_Coast_chalet_avg_price = North_Coast['Price'].astype(float).mean()
North_Coast_chalet_count = North_Coast.shape[0]


```

```python
import matplotlib.pyplot as plt

# البيانات
locations = ['Sokhna', 'Ras El Hekma', 'North Coast']
avg_prices = [chalet_sokhna_avg_price, ras_elhekma_chalet_avg_price, North_Coast_chalet_avg_price]
counts = [chalet_sokhna_count, ras_elhekma_chalet_count, North_Coast_chalet_count]

fig, ax1 = plt.subplots(figsize=(10, 6))

# رسم الأعمدة لمتوسط الأسعار
colors = ['tab:blue', 'tab:orange', 'tab:red']
ax1.set_xlabel('Location')
ax1.set_ylabel('Average Price (EGP)')
bars = ax1.bar(locations, avg_prices, color=colors, alpha=0.6)

# إضافة التسميات على الأعمدة
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f} EGP', va='bottom', ha='center', color='black', fontsize=10)

ax1.tick_params(axis='y')

# إعداد المحور الثاني لرسم عدد الشاليهات
ax2 = ax1.twinx()
ax2.set_ylabel('Count')
lines = ax2.plot(locations, counts, color='tab:green', marker='o', markersize=10, label='Count', linestyle='--')
ax2.tick_params(axis='y')

# إضافة التسميات على النقاط
for i, txt in enumerate(counts):
    ax2.annotate(txt, (locations[i], counts[i]), textcoords="offset points", xytext=(0,10), ha='center', color='green', fontsize=10)

# إضافة العناوين والتسميات
plt.title('Average Price and Count of Chalets by Location')

# إضافة وسوم للرسومات البيانية
ax1.legend(bars, ['Sokhna', 'Ras El Hekma', 'North Coast'], loc='upper left', title='Average Price')
ax2.legend(lines, ['Count'], loc='lower right', title='Count')

# عرض الرسم البياني
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# افتراض أن df هو DataFrame الأصلي الذي يحتوي على البيانات

# استخراج الصفوف التي تحتوي على 'Ain Sukhna' في عمود 'Location'
sokhna = df.loc[df['Location'].str.contains('Ain Sukhna', case=False, na=False)]

# استخراج أنواع الوحدات
unit_types = sokhna['Type'].unique()

# حساب متوسط الأسعار لكل نوع وحدة
avg_prices = []
for unit_type in unit_types:
    avg_price = sokhna[sokhna['Type'] == unit_type]['Price'].astype(float).mean()
    avg_prices.append(avg_price)

# إنشاء DataFrame للعرض
units_avg_price_df = pd.DataFrame({
    'Unit Type': unit_types,
    'Average Price': avg_prices
})

# قائمة الألوان لكل نوع وحدة
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'plum', 'gold', 'lightpink', 'lightgray']

# رسم بياني لمتوسط الأسعار لكل نوع وحدة
plt.figure(figsize=(10, 6))
bars = plt.bar(units_avg_price_df['Unit Type'], units_avg_price_df['Average Price'], color=colors[:len(unit_types)])

# إضافة الأسعار فوق الأشرطة
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.2f}', ha='center', va='bottom')

# إضافة legend
for bar, color, label in zip(bars, colors, units_avg_price_df['Unit Type']):
    bar.set_label(label)
plt.legend()

plt.xlabel('Unit Type')
plt.ylabel('Average Price')
plt.title('Average Prices of Units in Ain Sukhna')
plt.xticks(rotation=45)
plt.show()
```
